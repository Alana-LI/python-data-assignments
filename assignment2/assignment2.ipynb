```python
import seaborn as sns
import matplotlib.pyplot as plt 
def word_count(column):
    word_count = pd.Series(column).value_counts().sort_values(ascending=False)[0:5]  
    fig = plt.figure(figsize=(16,8))  
    x = word_count.index.tolist()  
    y = word_count.values.tolist()  
    sns.barplot(x, y, palette="BuPu_r")  
    plt.title('Author frequency top 5')  
    plt.ylabel('count')  
    sns.despine(bottom=True)  
    plt.show()

word_count(column)
```

```python
list = column #the list of words you have
dict_words_frequency={}
for n in list:
    dict_words_frequency[n]=list.count(n)
#dict_words_frequency
```

```python
f = open('title.txt','w')
for i in range(len(list)):
    f.write(list[i])
f.close()
```

```python
f=open('title.txt','r')
contents= f.read()
all_words = "".join(contents)
for ch in '\s+\.\!\/_,$%^*(+\"\')]+|[+——()?:【】“”‘’！，。’,':
    words = all_words.replace(ch," ")
```

```python
#words
```

```python
words = words.split()
```

```python
#words
```

```python
filepath = 'stopwords.txt'
stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()] 
```

```python
#stopwords
```

```python
processed_word_list = [] 
for word in words:
    word = word.lower() # in case they are not all lower cased
    if word not in stopwords:
        processed_word_list.append(word)
```

```python
#processed_word_list
```

```python
word_count = pd.Series(processed_word_list).value_counts().sort_values(ascending=False)[0:15]
word_count
```

# which words appeared most frequently?
* Trump,migrant and mexico are the keywords which appears most frequently, it was the time Trump said he wanted giant wall across Africa to stop the migrants getting into Europes'.

```python
import wordcloud
import numpy as np

def tag_cloud(text):
#     mask = np.array(Image.open('newspaper.png')) #set mask, you can change to the picture you like, but it must have a high color contrast
    wc = wordcloud.WordCloud(mode='RGBA',background_color='white',max_words=2000,stopwords=stopwords,max_font_size=300,random_state=42)
    wc.generate_from_text(' '.join(text))
    plt.figure(figsize=(12,12))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.title('word frequency top 15 tag cloud', loc='Center', fontsize=20)
    plt.show()
    return plt.show()

tag_cloud(processed_word_list)
```
