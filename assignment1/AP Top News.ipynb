```python
import requests
r = requests.get('https://apnews.com/apf-topnews')
```

```python
r.text
```

```python
import bs4
```

```python
mypage=bs4.BeautifulSoup(r.text)
mypage
```

```python
links=mypage.find_all('a',attrs={'class':'headline'})
links
```

```python
for i in range(len(links)):
    link=links[i]['href']
    print(link)
```

```python
Links=[]
for link in links:
    url='{0}{1}'.format('https://apnews.com',link['href'])
    Links.append(url)
Links
```

```python
headlines=mypage.find_all('h1')
headlines
```

```python
len(headlines)
```

```python
for i in range(50):
    headline=headlines[i].text
    print(headline)
```

```python
Headlines=[]
for headline in headlines:
    Headlines.append(headline.text)
Headlines
```

```python
authors=mypage.find_all('span',attrs={'class':'byline'})
```

```python
authors
```

```python
for i in range(len(authors)):
    author=authors[i].text
    print(author)
```

```python
Authors=[]
for author in authors:
    Authors.append(author.text)
Authors
```

```python
tags=mypage.find_all('a',attrs={'class':'HubTag'})
tags
```

```python
for i in range(len(tags)):
    tag=tags[i].text
    print(tag)
```

```python
Tags=[]
for tag in tags:
    Tags.append(tag.text)
Tags
```

```python
times=mypage.find_all('span',attrs={'class':'Timestamp'})
times
```

```python
for i in range(len(times)):
    time=times[i].text
    print(time)
```

```python
Times=[]
for time in times:
    Times.append(time.text)
Times
```

```python
places=mypage.find_all('p')
places
```

```python
for i in range(len(places)):
    place=places[i].text
    print(place)
```

```python
Places=[]
for place in places:
    Places.append(place.text.split('(')[0])
Places
```

```python
Z=[]
for i in range(0,49):
    z=[Headlines[i],Places[i],Links[i],Authors[i],Tags[i],Times[i]]
    Z.append(z)
```

```python
Z
```

```python
import csv
with open('AP Top News.csv','a',newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    header=['headline','place','link','author','tag','time']
    spamwriter.writerow(header)
    spamwriter.writerows(Z)
```
