```python
f1=open("trade-wars-news1.txt","r+")
content1=f1.read()
f1.close()
f2=open("trade-wars-news2.txt","r+")
content2=f2.read()
f2.close
f3=open("trade-wars-news3.txt","r+")
content3=f3.read()
f3.close()
f4=open("trade-wars-news4.txt","r+")
content4=f4.read()
f4.close()
f5=open("trade-wars-news5.txt","r+")
content5=f5.read()
f5.close()

```

```python
A = str.split(content1)
B = str.split(content2)
C = str.split(content3)
D = str.split(content4)
E = str.split(content5)
all_words = A + B + C + D + E
all_words
```

```python
stop_words=['the','and','of','a','in','to','on','for','that','is','The']
for i in range(len(stop_words)):
    while stop_words[i] in all_words:
        all_words.remove(stop_words[i])
all_words
```

```python
'the' in all_words
```

```python
all_words_dict={}
for i in all_words:
    all_words_dict[i]=all_words.count(i)
all_words_dict
```

```python
type(all_words_dict)
```

```python
sortedwordlist=sorted(all_words_dict.items(),key=lambda x:x[1],reverse=True)
sortedwordlist
```

```python
list=sortedwordlist
print(sortedwordlist[0:15])
```
